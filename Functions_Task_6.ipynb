1:Smart Parking Lot Management System

```python
# Function to manage parking lot
def manage_parking(capacity, logs):
    parked = 0 # Current parked vehicles
    peak = 0  # Peak parking usage
    # Process each log
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT":
            parked -= 1
        # Track peak usage
        if parked > peak:
            peak = parked
    print("Currently Parked Vehicles:", parked)
    # Check parking status
    if parked > capacity:
        print("Parking Status: Capacity Exceeded ⚠")
    else:
        print("Parking Status: Available")
    print("Peak Parking Usage:", peak)
# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
# Call function
manage_parking(capacity, vehicle_logs)
```

2:Online Food Delivery Time Estimator

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):
    # Base time: 5 minutes per km
    time = distance * 5
    # Add delay for traffic
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    # Low traffic → no extra time
    # Add delay for weather
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 15
    # Clear weather → no extra time
    print("Estimated Delivery Time:", time, "minutes")
# Sample Input
distance_km = 8
traffic_level = "High"
weather_condition = "Rainy"
# Call function
estimate_delivery_time(distance_km, traffic_level, weather_condition)
```

3:Movie Theatre Seat Occupancy Analyzer

```python
# Function to analyze seat occupancy
def analyze_seats(total_seats, booked_seats):
    # Count booked seats using loop
    booked_count = 0
    for seat in booked_seats:
        booked_count += 1
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")
# Sample Input
total = 200
booked = [1] * 150   # 150 seats booked
# Call function
analyze_seats(total, booked)
```

4:Cloud Server Load Classification System

```python
# Function to classify server load
def classify_server_load(cpu_readings):
    total = 0   # Variable to store total CPU usage
    # Calculate total using loop
    for value in cpu_readings:
        total += value
    # Calculate average CPU load
    average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")
    # Classify server status
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
# Sample Input
cpu_data = [45, 60, 70, 85, 90]
# Call function
classify_server_load(cpu_data)
```

5:Smart Classroom Resource Usage Monitor

```python
# Function to check resource usage
def check_resource_usage(resources):
    overused = []   # List to store overused resources
    # Loop through dictionary
    for resource, hours in resources.items():
        # If usage is more than 8 hours → overuse
        if hours > 8:
            overused.append(resource)
    # Print overused resources
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
# Call function
check_resource_usage(resource_usage)
```

6: Online Event Registration Capacity Controller

```python
# Function to manage event registrations
def manage_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    # Check registrations against capacity
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    # Print results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
# Sample Input
event_capacity = 100
registrations = 105
# Call function
manage_registrations(event_capacity, registrations)
```
