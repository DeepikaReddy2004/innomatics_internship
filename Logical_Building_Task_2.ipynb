Problem 1: Unique Words in a Sentence

```python
s=input("Enter sentence:")
s=s.split()
unique_words=set(s)
print(unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_name=""
highest_salary=0
for name,salary in employees.items():
  if salary>highest_salary:
    highest_salary=salary
    highest_name=name
print(f'Highest Salary :{highest_name}-{highest_salary}')
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
min=numbers[0]
max=numbers[0]
for i in numbers:
  if i>max:
    max=i
  if i<min:
    min=i
print("Max:",max)
print("Min",min)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
  if i>1000:
    count+=1
print("products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present=0
for i in attendance:
  if i=="P":
    present+=1
attendance_percentage=present/len(attendance)*100
print("Attendance Percentage:",attendance_percentage)
```

Problem 6: Remove Duplicate Phone Numbers

Problem 7: Count Character Frequency

```python
text=input("Enter a string:").strip()
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print(char_count)
```

Problem 10: Calculate Average Marks

```python
marks=[80,90,86,95,98,94]
total=0
for i in marks:
  total+=i
average_marks=total/len(marks)
print("Average Marks:",average_marks)
```
