1: Social Media – Post Engagement Analyzer

```python
# Function to analyze post engagement
def analyze_post(likes_list):
    total_likes = 0   # Variable to store total likes
    # Calculate total likes using loop
    for likes in likes_list:
        total_likes += likes
    print("Total Likes:", total_likes)

    # Check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
# Sample input (likes from different users or platforms)
likes_data = [200, 300, 180, 400]
# Call function
analyze_post(likes_data)
```

3: Agriculture – Rainfall Adequacy Checker

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):
    total = 0   # Variable to store total rainfall
    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total += rain
    # Calculate average rainfall
    average = total / len(rainfall_data)
    print("Average Rainfall:", int(average))
    # Check rainfall status
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
# Sample rainfall data
seasonal_rainfall = [70, 75, 68, 80, 67]
# Required rainfall level
required = 70

# Call function
check_rainfall(seasonal_rainfall, required)
```

4: Social Media – Duplicate Account Detection

```python
# Function to check duplicate usernames
def check_duplicates(usernames):
    unique_users = set()   # Set to store unique usernames
    duplicate_found = False
    # Loop through usernames
    for user in usernames:
        # If username already in set → duplicate
        if user in unique_users:
            duplicate_found = True
            break
        # Otherwise add to set
        unique_users.add(user)
    # Print result
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Sample usernames list
users = ["Deepika", "Pranathi", "Navadeep Reddy", "Deepika"]

# Call function
check_duplicates(users)
```

6: Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crops
def filter_premium_crops(prices):
    premium_crops = []   # List to store premium prices
    # Loop through prices
    for price in prices:
        # Check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)
    # Print premium crops
    print("Premium Crops:", premium_crops)
# Example crop prices
crop_prices = [1500, 2500, 1800, 3200, 1900]

# Call function
filter_premium_crops(crop_prices)
```

9: E-Learning – Student Attendance Eligibility System

```python
# Function to check exam eligibility
def check_exam_eligibility(attendance_list):
    present_count = 0   # To count "P"
    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1
    # Calculate attendance percentage
    percentage = (present_count / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)
    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
# Example attendance data
attendance = ["P", "P", "A", "P", "P"]   # 4 out of 5 present → 80%

# Call function
check_exam_eligibility(attendance)
```

10: Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill
def calculate_bill(units):
    bill = 0
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Classify usage
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
# Example units consumed
units_consumed = 250

# Call function
total_bill, usage_status = calculate_bill(units_consumed)

print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
