1. User Login Check

```python
username=input("Enter username:")
password=input("Enter Password:")
if username=="admin" and password=="1234":
  print("Login Successful")
else:
  print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    pass_count+=1
  else:
    fail_count+=1
print(f'Number of students passed is {pass_count} and number of students failed is {fail_count}')
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
res=[]
for i in names:
   n=i.strip().lower()
   res.append(n)
print(res)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  l=len(i)
  print(f'message {i} has length {l}')
  if l>10:
    print(f'Long message {i}')
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
    count+=1
print(f"Number of ERROR entries: {count}")
```
