## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
"Love" and "love" are treated as different tokens in NLP if case is not normalized.
To avoid duplication and inconsistency, text is converted to lowercase.

### 2. What happens if stopwords are not removed?
Stopwords increase noise in the dataset and add unnecessary words.
This may reduce model performance and increase computation.

### 3. When removing stopwords can be harmful?
- In sentiment analysis: "not happy" becomes "happy"
- In question answering: removing words like "what", "how" removes meaning

### 4. Stemming vs Lemmatization
Stemming reduces words to root form roughly (running → runn).
Lemmatization converts words into meaningful base form (running → run).

```python
import re
from collections import Counter
```

## Task 2: Advanced Preprocessing Function

```python
def preprocess_text(text):
    # handle invalid or empty input
    if not isinstance(text, str) or text.strip() == "":
        return []

    # convert to lowercase
    text = text.lower()

    # remove urls and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # remove numbers
    text = re.sub(r'\d+', '', text)

    # handle repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)

    # remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # tokenize
    tokens = text.split()

    # remove short words but keep 'no', 'not'
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not']:
            cleaned_tokens.append(word)

    return cleaned_tokens
```

## Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens = preprocess_text(s)
    clean_sentence = " ".join(tokens)

    print("\n-------------------------")
    print("Original Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
```

## Task 4: Token Analytics

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total if total > 0 else 0
    return total, unique, round(avg_len, 2)

for s in sentences:
    tokens = preprocess_text(s)
    total, unique, avg = token_analysis(tokens)

    print("\nSentence:", s)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)
```

## Task 5: Frequency Analysis

```python
all_tokens = []

for s in sentences:
    all_tokens.extend(preprocess_text(s))

freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(freq.most_common()[:-6:-1])
```

## Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
