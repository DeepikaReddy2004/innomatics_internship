1: Smart Light Controller

```python
# Smart Light Controller
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(self.light_name + " is " + self.status)
# Taking input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")
# Creating object
light = SmartLight(light_name)
# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")
# Display result
light.display_status()
```

2: Employee ID Card System

```python
# Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
# Creating object
emp = Employee(name, emp_id, department)
# Displaying ID card
emp.display_id_card()
```

3: Mobile Contact Record

```python
# Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
# Taking input
name = input("Contact Name: ")
phone = input("Phone Number: ")
# Creating object
c1 = Contact(name, phone)
# Display contact information
c1.display_contact()
```

 4: Product Price Tag Generator

```python
# Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))
# Taking input
name = input("Product Name: ")
price = input("Price: ")
# Creating object
p1 = Product(name, price)
# Printing price tag
p1.print_price_tag()
```

5: Movie Rating Display System

```python
# Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_movie(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")
# Taking input
name = input("Movie Name: ")
rating = input("Rating: ")
# Creating object
m1 = Movie(name, rating)
# Displaying movie details
m1.display_movie()
```

6: Delivery Address Manager

```python
# Delivery Address Manager
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
# Taking input
name = input("Customer Name: ")
address = input("Address: ")
# Creating object
d1 = Delivery(name, address)
# Displaying delivery details
d1.display_details()
```
