1: Mobile Recharge Validation System

```python
# List of valid recharge plans
valid_plans = [199, 299, 399, 599]

# Keep asking user until valid recharge is entered
while True:
    # Take recharge amount as input
    amount = int(input("Enter recharge amount: "))
    # Check minimum recharge amount
    if amount < 50:
        print("Amount must be at least ₹50")
    # Check if amount is in valid plans
    elif amount not in valid_plans:
        print("Invalid recharge plan")
    # If both conditions are satisfied
    else:
        print("Recharge Successful")
        break   # Stop the loop
```

2: Inventory Reorder Alert System

```python
# Function to check stock
def check_inventory(products):
    # Loop through dictionary
    for item, stock in products.items():
        # If stock is less than 15
        if stock < 15:
            print(f"{item} : Reorder Alert ")
        # Otherwise stock is sufficient
        else:
            print(f"{item} : Stock OK")
# Sample input dictionary
products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 18
}

# Call function
check_inventory(products)
```

3: Student Result Processing System

```python
# Function to process student marks
def check_result(marks):
    total = 0   # Variable to store total marks
    # Calculate total using loop
    for m in marks:
        total += m
    # Calculate average
    average = total / len(marks)
    # Check pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
# Sample marks list
student_marks = [60, 55, 40, 70]

# Call function and print result
result = check_result(student_marks)
print("Result:", result)
```

4: Cab Fare Estimator with Retry Option

```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)
    # Add 25% extra if peak hour
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
# Retry option using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        break
```

5: Employee Attendance Eligibility

```python
# Function to check attendance eligibility
def check_eligibility(attendance):
    present_count = 0   # To count number of "P"
    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_count += 1
    # Calculate percentage
    percentage = (present_count / len(attendance)) * 100
    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
# Sample attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function and print result
result = check_eligibility(attendance_list)
print("Attendance Result:", result)
```

6: Password Strength Checker

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    # Check each character in password
    for ch in password:
        # Check if character is digit
        if ch.isdigit():
            has_digit = True
        # Check if character is special (@ # $)
        if ch in special_chars:
            has_special = True
    # Check all conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
# Take input from user
user_password = input("Enter password: ")

# Check password strength
result = check_password(user_password)
print(result)
```
